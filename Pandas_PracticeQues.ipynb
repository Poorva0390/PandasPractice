```python
# Get unique items from series_1

ser_1 = pd.Series([1,4,2,7,8])
ser_2 = pd.Series([5,7,3,1,9])

print(ser_1[~ser_1.isin(ser_2)])
```

```python
# Get unique item from both the series

series_union = pd.Series(np.union1d(ser_1,ser_2))
series_common = pd.Series(np.intersect1d(ser_1,ser_2))
#print(series_union)
print(series_union[~series_union.isin(series_common)])
```

```python
#Frequency of count unique items of a series
import pandas as pd
ser_3 = pd.Series([1,1,3,5,3,2,5,3,2,1,7,8,9,5,7])
print(ser_3.value_counts())
```

```python
#Replace everything as 'Other' except top 2 most frequent values

ser_3[~ser_3.isin(ser_3.value_counts().index[:2])]= 'Other'
print(ser_3)
```

```python
#Get the day, month, week number,day of year, day of a week from a series of date strings

set_dateStr1 = pd.Series(['01 Jan 2019','30 Jun 2015','29 Aug 2017'])

from dateutil.parser import parse

ser_ts= set_dateStr1.map(lambda x : parse(x))
print('Date:' ,ser_ts.dt.day.tolist())
print('Month:' ,ser_ts.dt.month.tolist())
print('Year:' ,ser_ts.dt.year.tolist())
print('WeekNumber:' ,ser_ts.dt.week.tolist())
print('Day Number:' ,ser_ts.dt.dayofyear.tolist())
#print('Day of week:' ,ser_ts.dt.weekday_name.tolist())


```

```python
#Convert list of nested dictionary into Pandas dataframe

list_nestedDictionary = [
                            {
                                "Student" : [{"Exam" : 90 , "Grade" : "a"},
                                           {"Exam" : 88 , "Grade" : "b"}, 
                                             {"Exam" : 100 , "Grade" : "a"}
                                            ],
                                "Name" : "Poorva Yadav"
                                
                            },
    
                            {
                                "Student" : [{"Exam" : 70 , "Grade" : "c"},
                                           {"Exam" : 60 , "Grade" : "d"} 
                                            ],
                                "Name" : "Poorva_Y"
                                
                            }
]

print(list_nestedDictionary)

```

```python
#Adding dict values to rows

import pandas as pd
rows_to_dict=[]

for data in list_nestedDictionary:
    data_row = data['Student']
    time = data['Name']
    
    for row in data_row:
        row['Name'] = time
        rows_to_dict.append(row)
        
dfr = pd.DataFrame(rows_to_dict)
print(dfr)
        
```

```python
# Dataframe apply with arguments

import pandas as pd

def sum(x,y,z,m):
    return (x+y+z)*m

df=pd.DataFrame({'A' : [1,2],'B' : [10,20]})
df1 = df.apply(sum,args=(1,2,10))
print(df1)
```
